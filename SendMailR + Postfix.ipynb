```r
mensagem <- function(nome, email){
    msg <- mime_part(paste(msg_pt1, nome, msg_pt2, email, msg_pt3, sep=''))
    msg[["headers"]][["Content-Type"]] <- "text/html"
    return(msg)
}
```

#### Configurações

```r
from <- "projetosifilis@lais.huol.ufrn.br"
subject <- "Apoiadores - Email Institucional - #TesteTrateCure"
```

#### Usuarios

```r
usuarios <- read.csv2("usuarios.csv", header=T, sep=";")
```

```r
str(usuarios)
```

#### Função de Envio

```r
email <- function(usr){
    name <- usr["Nome"] 
    email <- usr["Email.Criado"]
    to <- usr["Email.Pessoal"]
    
    response <- sendmail(from=from,to=to,subject=subject,msg=mensagem(name, email))
    
    paste(name, response[1], sep="--")
}
```

#### Enviando

```r
apply(usuarios, 1, email)
```
